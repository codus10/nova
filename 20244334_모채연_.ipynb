**미션 1. 한글 CSV 파일 불러오기 및 글자 깨짐 해결**

* **상황:** 공공데이터포털에서 다운받은 `seoul_population.csv` 파일을 `df`라는 변수로 불러오고 싶습니다. 그런데 한글이 깨져서 나옵니다.
* **나의 프롬프트:** > "파이썬 판다스를 써서 `seoul_population.csv` 파일을 읽어오려고 하는데 한글이 자꾸 깨져. 글자 깨짐 현상을 해결해서 `df`라는 데이터프레임으로 불러오는 코드를 짜줘."

**미션 2. 복잡한 다중 조건 필터링 (AND, OR 연산)**

* **상황:** `students` 데이터프레임에서 전공(`major`)이 '컴퓨터공학'이면서(AND), 학점(`gpa`)이 4.0 이상인 학생들만 추출하여 새로운 데이터프레임 `honor_students`를 만들고 싶습니다.
* **나의 프롬프트:** > "파이썬 판다스를 사용해서 students 데이터프레임에서 전공이 '컴퓨터공학'이고 학점이 4.0 이상인 학생들만 추출해서 honor_students라는 새로운 데이터프레임을 만들어줘.
"


````python
# @title
import pandas as pd
import random
import numpy as np

# 시드 고정
random.seed(42)
np.random.seed(42)

# 학생 수
n = 100

# 학번: 8자리, 앞 4자리는 연도(예: 2026)
ids = [f"2026{str(i).zfill(4)}" for i in range(1, n+1)]

# 전공: "AI", "CS", "EE" 랜덤 배치
majors = [random.choice(["AI", "CS", "EE"]) for _ in range(n)]

# 성별: "M", "F" 랜덤 배치
sexes = [random.choice(["M", "F"]) for _ in range(n)]

# 학점: 평균 3.5, 표준편차 0.5, 범위 1.0 ~ 4.5
gpas = np.random.normal(loc=3.5, scale=0.5, size=n)
gpas = np.clip(gpas, 1.0, 4.5)

# 소수점 2자리로 반올림
gpas = [round(g, 2) for g in gpas]

# 데이터 프레임 생성
students = pd.DataFrame({
    "id": ids,
    "major": majors,
    "sex": sexes,
    "gpa": gpas
})

print(students.head())  # 앞부분 확인
````

4. **결과 확인:** 살아남은 1등급 젊은 승객(`vip_survivors`)은 총 몇 명인지 화면에 출력합니다. (힌트: 데이터프레임의 행 개수를 세면 됩니다.)

````python
import pandas as pd
import seaborn as sns

# 타이타닉 데이터셋 불러오기
titanic = sns.load_dataset('titanic')

# 데이터 확인
print(titanic.head())
````

````python
print("데이터 크기:", titanic.shape)       # (행 개수, 열 개수)
print("\n데이터 정보:")
print(titanic.info())
````

````python
vip_survivors = titanic[(titanic['pclass'] == 1) &   # 1등석
                        (titanic['age'] < 30) &      # 30세 미만
                        (titanic['survived'] == 1)]
````

````python
print("\nVIP 생존자 수:", len(vip_survivors))   # 조건에 맞는 VIP 생존자 수 출력
print("\nVIP 생존자 데이터 샘플:")
print(vip_survivors.head())
````
