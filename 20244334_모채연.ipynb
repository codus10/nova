## 🎯 PART 1. 핵심 개념 확인

*판다스(Pandas)를 이용해 데이터를 요리하는 기본 문법과 용어들을 확인합니다.*

1. 표 형태의 데이터프레임에서 하나의 열(컬럼)만 떼어내어 가져올 때는 데이터프레임 이름 뒤에 **( 대괄호 )** 를 쓰고 그 안에 열 이름을 적습니다. (예: `df['이름']`)
2. 하나의 열을 가져오면 1차원 형태인 시리즈(Series)가 되지만, **두 개 이상** 의 열을 한 번에 가져오려면 대괄호를 두 번 겹쳐 쓴 `df[ ( 리스트 ) ]` 형태로 리스트를 넣어주어야 합니다.
3. 파이썬 기본 문법에서 '같다'는 `==`입니다. 그렇다면 '크거나 같다(이상)'를 의미하는 비교 연산자 기호는 **( >= )** 입니다.
4. 판다스에서 특정 조건에 맞는 데이터만 걸러내는 방식을 **( Boolean Indexing ) Indexing** 이라고 부릅니다. 조건이 참(True)인 행만 남기는 원리입니다.
5. 판다스 필터링에서 두 가지 조건이 **모두 참이어야 할 때(AND)** 는 일반 파이썬의 `and` 대신 기호 **( & )**(앰퍼샌드)를 사용해야 해요.
6. 판다스 필터링에서 두 가지 조건 중 **하나만 참이어도 될 때(OR)** 는 일반 파이썬의 `or` 대신 기호 **( | )**(파이프, 보통 Enter 키 위)를 사용해야 돼요.
7. 다중 조건을 사용할 때는 연산자 우선순위 때문에 헷갈리지 않도록 반드시 각 조건을 **( 괄호() )** 로 묶어주어야 에러가 나지 않습니다.
8. 특정 문자가 포함된 데이터를 찾을 때, 엑셀의 '텍스트 포함' 필터와 같은 역할을 하는 판다스의 문자열 메서드는 `str.( contains )('찾을단어')` 입니다.
9. 데이터를 특정 열을 기준으로 정렬할 때 사용하는 판다스의 대표적인 메서드는 `( sort_values() )` 입니다.
10. 데이터를 정렬할 때 기본값은 오름차순(작은 것부터 큰 것 순)입니다. 이를 **내림차순(큰 것부터 작은 것 순)** 으로 바꾸려면 정렬 함수 안에 `( ascendinf ) = False` 라는 옵션을 주어야 합니다.
11. 필터링이나 정렬을 하고 나면 행 번호(인덱스)가 뒤죽박죽 섞이게 됩니다. 이를 0부터 다시 깔끔하게 번호를 매겨주는 메서드는 `( reset_index() )` 입니다.

**미션 3. 다중 조건 필터링 (AND 조건)**

* **상황:** `sales` 데이터프레임에서 카테고리(`category`)가 '전자기기' 이면서(AND) 매출액(`revenue`)이 100만 원 이상인 '효자 상품' 행만 뽑아내고 싶습니다.
* **나의 프롬프트:** > "판다스로 sales 데이터프레임에서 category가 '전자기기'이고 revenue가 1000000 이상인 행만 필터링해서 hit_products라는 변수에 저장하는 코드를 짜줘."


````python
import pandas as pd

# 예시 데이터프레임 생성
data = {
    'product': ['노트북', '스마트폰', '책상', 'TV', '청소기'],
    'category': ['전자기기', '전자기기', '가구', '전자기기', '전자기기'],
    'revenue': [1500000, 800000, 500000, 2000000, 1200000]
}

sales = pd.DataFrame(data)

# category가 '전자기기'이고 revenue가 1000000 이상인 행만 필터링
hit_products = sales[(sales['category'] == '전자기기') & (sales['revenue'] >= 1000000)]

print(hit_products)
````

**미션 4. 텍스트 포함 조건으로 찾기 (`str.contains`)**

* **상황:** `users` 데이터프레임의 주소(`address`) 열에 '서울특별시'라는 글자가 포함된 사람들만 전부 찾아내고 싶습니다.

````python
import pandas as pd

# 예시 데이터프레임 생성
data = {
    'name': ['홍길동', '김철수', '이영희', '박민수'],
    'address': ['서울특별시 강남구', '부산광역시 해운대구', '서울특별시 마포구', '광주광역시 북구']
}

users = pd.DataFrame(data)

# address 열에 '서울특별시'가 포함된 행만 필터링
seoul_users = users[users['address'].str.contains('서울특별시', na=False)]

print(seoul_users)
````

## 🚀 PART 4. [심화 문제] VIP 타겟 마케팅 대상자 추출하기

*선택, 필터링, 정렬을 하나로 묶어 실무에서 가장 많이 하는 데이터 가공 작업을 수행해 봅니다.*

**[비즈니스 시나리오 및 요구사항]**
당신은 온라인 쇼핑몰의 마케터입니다. 신상품 VIP 초대장을 보낼 고객 명단을 뽑아야 합니다. `mall_data`라는 데이터프레임이 준비되어 있습니다. AI에게 아래 로직을 순서대로 설명하여 최종 결과물(`vip_list`)을 만드는 단 한 줄(또는 두 줄)의 코드를 받아내세요.

1. **필터링:** 회원 등급(`level`)이 'Gold' 이거나 'VIP'인 사람만 먼저 추려냅니다.
2. **필터링(AND):** 그중에서 이번 달 구매 금액(`monthly_spend`)이 50만 원 이상인 사람만 남깁니다.
3. **열 선택:** 분석에 필요 없는 개인정보는 빼고, '고객ID', '이름', 'monthly_spend' 열만 남깁니다.
4. **정렬:** 초대장을 구매 금액이 높은 사람부터 순서대로 보내기 위해, 'monthly_spend'를 기준으로 내림차순 정렬합니다.

* **✍️ 위 모든 조건을 한 번에 처리하기 위해 AI에게 내린 나의 구체적인 프롬프트:**
 판다스로 `mall_data` 데이터프레임에서
1) `level`이 'Gold' 또는 'VIP'인 고객 중
2) `monthly_spend`가 500000 이상인 행만 필터링하고,
3) '고객ID', '이름', 'monthly_spend' 열만 선택한 뒤
4) `monthly_spend` 기준으로 내림차순 정렬하여
`vip_list`라는 변수에 저장하는 코드를 짜줘.




* **💻 AI가 짜준 판다스 코드 완성본 적어보기:**

vip_list = mall_data[
    ((mall_data['level'] == 'Gold') | (mall_data['level'] == 'VIP'))
    & (mall_data['monthly_spend'] >= 500000)
][['고객ID', '이름', 'monthly_spend']].sort_values(by='monthly_spend', ascending=False)



* **✨ 스스로 생각하기:** 만약 위 과정이 끝난 후, 데이터프레임 제일 왼쪽의 인덱스 번호(행 번호)가 4, 19, 25 처럼 뒤죽박죽되어 있다면, 코드를 더 깔끔하게 만들기 위해 마지막에 어떤 메서드를 추가로 붙여주는 것이 좋을까요?
👉 `( reset_index(drop=True) )`

````python
import pandas as pd

# 예시 데이터프레임 생성
data = {
    '고객ID': [101, 102, 103, 104, 105],
    '이름': ['홍길동', '김철수', '이영희', '박민수', '최수진'],
    'level': ['Silver', 'Gold', 'VIP', 'Gold', 'Bronze'],
    'monthly_spend': [200000, 600000, 1200000, 800000, 300000]
}

mall_data = pd.DataFrame(data)

# VIP 명단 추출
vip_list = mall_data[
    ((mall_data['level'] == 'Gold') | (mall_data['level'] == 'VIP'))
    & (mall_data['monthly_spend'] >= 500000)
][['고객ID', '이름', 'monthly_spend']]\
.sort_values(by='monthly_spend', ascending=False)\
.reset_index(drop=True)

print(vip_list)
````
