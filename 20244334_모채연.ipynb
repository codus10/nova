2. 데이터의 형태를 의미하는 자료형 중, `10`, `-5`와 같은 숫자는 **(정수)**(Integer)이라고 부릅니다.

5. 파이썬의 리스트는 순서를 세기 시작할 때 1이 아니라 **(0)** 부터 번호를 매깁니다. 이 번호를 인덱스(Index)라고 합니다.

7. 리스트 안에서 콜론(`:`)을 사용하여 원하는 구간의 데이터만 잘라내어 새로운 리스트로 만드는 작업을 **(슬라이싱)**(Slicing)이라고 합니다.
8. 리스트의 **맨 마지막**에 새로운 데이터를 하나 추가할 때 사용하는 메서드는 `( append( ))`입니다.
9. 리스트의 **특정 위치(인덱스)** 를 지정하여 데이터를 끼워 넣을 때 사용하는 메서드는 `(insert())`입니다.
10. 리스트에서 **특정 데이터 값을 찾아서 삭제** 할 때 사용하는 메서드는 `remove()`이고, **특정 인덱스의 값을 뽑아내며 삭제**할 때는 `(pop())`를 사용합니다.
11. 현재 리스트 안에 데이터가 총 몇 개 들어있는지 **길이(요소의 개수)** 를 확인할 때 사용하는 내장 함수는 `(len())`입니다.
12. 문자열이나 리스트 안의 데이터를 **가나다순 또는 오름차순** 으로 정렬해 주는 리스트 메서드는 `(sort())`입니다.

나만의 리스트 (업데이트)

````python
my_favorites = [
    "인셉션", "어벤져스", "타이타닉", "라라랜드", "기생충",
    "조커", "인터스텔라", "매트릭스", "해리포터", "극한직업"
]
print("정렬 전 리스트:", my_favorites)
````

리스트 가나다순 정렬 (sort)

````python
my_favorites.sort()
print("정렬 후 리스트:", my_favorites)
````

리스트 개수 세기 (len)

````python
count = len(my_favorites)
print("리스트에 들어있는 항목 개수:", count)
````
