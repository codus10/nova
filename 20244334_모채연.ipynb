## 🎯 PART 1. 개념 확인

*프로그램의 흐름을 제어하는 핵심 문법과 논리 연산의 개념을 확인합니다.*

1. 프로그램이 위에서 아래로만 실행되지 않고, 조건에 따라 갈래를 나누거나 특정 구간을 반복하도록 흐름을 통제하는 문법을 총칭하여 **(제어문)**(Control Flow)이라고 부릅니다.
2. 조건문의 뼈대가 되는 자료형으로, 오직 `True`(참) 또는 `False`(거짓) 두 가지 값만 가질 수 있는 자료형을 **(불리언)**(Boolean) 타입이라고 합니다.
3. 파이썬에서 두 값이 '같은지' 비교할 때는 `=`가 아니라 **(==)** 연산자를 사용하고, '다른지' 비교할 때는 `!=` 연산자를 사용합니다.
4. 여러 개의 조건을 결합할 때 사용하는 논리 연산자로, 양쪽 조건이 모두 참일 때만 True를 반환하는 연산자는 **(and)**입니다. (예: `A ___ B`)
5. 파이썬은 중괄호 `{}` 대신 띄어쓰기 4칸을 사용하여 "이 코드는 if문(또는 for문)에 속해 있다"는 블록(Block) 구조를 인식합니다. 이를 **(들여쓰기)**(Indentation)이라고 부릅니다.
6. `if` 조건이 거짓일 때 다음으로 검사할 조건을 추가로 제시하려면 **(elif)** 키워드를 사용합니다. (`else if`의 줄임말)
7. 횟수가 정해져 있거나, 리스트 같은 순차적인 데이터(Sequence)의 요소를 처음부터 끝까지 하나씩 꺼내며 반복할 때는 **(for문)**문을 사용합니다.
8. `for`문과 자주 함께 쓰이며, 시작 숫자부터 끝 숫자 직전까지 연속된 정수 배열을 생성해 주는 내장 함수는 **(range())**입니다. (예: `___(1, 5)` -> 1, 2, 3, 4)
9. 특정 조건이 참(True)인 동안에는 횟수에 상관없이 무한히 반복을 수행하는 문법은 **(while문)**문입니다.
10. 반복문(For, While)을 실행하던 도중, 특정 조건을 만족했을 때 **반복문을 즉시 강제 종료**하고 빠져나오는 키워드는 **(break)**입니다.
11. 반복문을 종료하지는 않지만, **현재 차례의 남은 코드를 건너뛰고** 다음 반복 차례로 곧바로 넘어가게 하는 키워드는 **(continue)**입니다.

## 🎯 PART 2. 코드 결과 예측하기

*코드가 어떤 순서로 실행될지, 조건에 따라 어떤 분기(Branch)를 탈지 머릿속으로 추적해 보세요.*

**문제 1. if-elif-else의 흐름 (조건의 순서)**
다음 코드를 실행했을 때 화면에 출력될 등급은 무엇일까요?

```python
score = 85
if score >= 80:
    print("B 등급")
elif score >= 85:
    print("B+ 등급")
else:
    print("C 등급")

```

* **예상 결과:** B등급 (이유: 위에서부터 조건을 검사할 때), if-elif-else 구문은 위에서부터 차례대로 조건을 검사합니다.
첫 번째 조건 if score >= 80: → score = 85이므로 참(True)입니다. 따라서 바로 "B 등급"을 출력하고, 이후의 elif나 else는 검사하지 않고 건너뜁니다.
비록 elif score >= 85: 조건도 참이지만, 이미 첫 번째 if에서 실행이 끝났기 때문에 도달하지 않습니다. 즉, 조건문은 위에서부터 순차적으로 검사하며, 한 번 실행되면 아래 조건은 무시된다는 점이 핵심입니다.


**문제 2. for문과 range()의 결합**
`range(시작, 끝, 스텝)`의 원리를 생각하며 출력될 숫자를 순서대로 적어보세요.

```python
for i in range(1, 6, 2):
    print(i, end=" ")

```

* **예상 결과:** 1 3 5 (힌트: `end=" "`는 줄바꿈 대신 띄어쓰기를 하라는 의미입니다.)

이유:
range(1, 6, 2)은 시작=1, 끝=6(직전까지), 스텝=2를 의미합니다.
따라서 생성되는 수열은 1 → 3 → 5입니다. (6은 포함되지 않음)
print(i, end=" ") 때문에 줄바꿈 대신 띄어쓰기로 출력됩니다.

**문제 3. 반복문을 이용한 누적 합계 (Accumulator Pattern)**
알고리즘의 가장 기본이 되는 누적 합계 패턴입니다. `total`의 최종 값은?

```python
total = 0
prices = [100, 200, 300]
for p in prices:
    total = total + p
print(total)

```

* **예상 결과:** 600
이유:

total = 0에서 시작합니다.

리스트 prices = [100, 200, 300]의 각 요소를 순서대로 p에 대입하며 누적합니다.

계산 과정:

total = 0 + 100 → 100

total = 100 + 200 → 300

total = 300 + 300 → 600

따라서 최종적으로 print(total)은 600을 출력합니다.


**문제 4. while문과 무한 루프 제어 (break)**
`while True:`는 무한 루프를 만듭니다. 코드가 언제 멈출지 예측해 보세요.

```python
count = 5
while True:
    count -= 1
    if count == 2:
        break
print(count)

```

* **예상 결과:** 2

이유:

count = 5에서 시작합니다.

while True: → 무한 루프이므로 내부에서 break가 실행될 때까지 계속 반복합니다.

반복 과정:

첫 번째 반복: count = 5 - 1 → 4 → 조건 count == 2는 거짓 → 계속 진행

두 번째 반복: count = 4 - 1 → 3 → 조건 거짓 → 계속 진행

세 번째 반복: count = 3 - 1 → 2 → 조건 참 → break 실행 → 반복 종료

따라서 루프가 끝난 뒤 print(count)은 2를 출력합니다.

**문제 5. 반복 건너뛰기 (continue)**
이 코드는 1부터 5까지 반복하면서 어떤 숫자들만 출력할까요?

```python
for n in range(1, 6):
    if n % 2 == 0:
        continue
    print(n, end=" ")

```

* **예상 결과:** 1 3 5

이유: range(1, 6)은 1부터 5까지 반복합니다.

각 숫자 n에 대해 if n % 2 == 0: 조건을 검사합니다.

짝수(2, 4)일 경우 continue가 실행되어 출력을 건너뛰고 다음 반복으로 넘어갑니다.

홀수(1, 3, 5)일 경우 조건이 거짓이므로 print(n, end=" ")가 실행됩니다.

## 🚀 PART 4. [심화 문제] 스무고개 숫자 맞추기 (Up & Down 게임)

*지금까지 배운 변수, 조건문(if), 반복문(while), 입출력(input, print), 그리고 난수 생성(random)을 모두 융합하여 미니 프로그램을 완성하는 종합 알고리즘 실습입니다.*

**[알고리즘 설계도 (Flowchart)]**

1. 컴퓨터가 `random.randint(1, 50)`을 사용해 1부터 50 사이의 정답 숫자를 하나 생성하여 변수에 저장합니다.
2. 사용자가 정답을 맞힐 때까지 **계속해서(while True)** 숫자를 입력(`input`)받습니다. (입력받은 값은 정수형 `int`로 변환해야 합니다.)
3. 만약 사용자가 입력한 숫자가 정답보다 **크다면**, "Down!"을 출력합니다.
4. 만약 사용자가 입력한 숫자가 정답보다 **작다면**, "Up!"을 출력합니다.
5. 만약 사용자가 입력한 숫자와 정답이 **같다면**, "정답입니다!"를 출력하고 **반복문을 즉시 종료(break)** 합니다.

**[실습 미션]**
게임을 만들기 위하여 AI에게 적절한 프롬프트를 전달하여 AI가 짜준 코드를 Colab에 붙여넣고 직접 게임을 플레이하며 로직이 정상적으로 작동하는지 검증합니다.

* **💻 AI가 작성해 준 코드를 검토하고 실행하기.**

나의 프롬프트 예시

"파이썬으로 1부터 50 사이의 난수를 random.randint(1, 50)으로 생성하고, 사용자가 input()으로 숫자를 입력할 때마다 정답과 비교하여 크면 'Down!', 작으면 'Up!'을 출력하고, 같으면 '정답입니다!'를 출력한 뒤 반복을 종료하는 코드를 작성해줘."

코드 예시

import random

# 1. 정답 숫자 생성
answer = random.randint(1, 50)

# 2. 무한 반복
while True:
    guess = int(input("숫자를 입력하세요 (1~50): "))
    
    # 3. 조건 비교
    if guess > answer:
        print("Down!")
    elif guess < answer:
        print("Up!")
    else:
        print("정답입니다!")
        break

실행 흐름

컴퓨터가 1~50 사이의 난수를 하나 뽑습니다.

사용자가 숫자를 입력합니다.

입력값이 정답보다 크면 "Down!", 작으면 "Up!"을 출력합니다.

정답을 맞히면 "정답입니다!"를 출력하고 게임이 종료됩니다.
